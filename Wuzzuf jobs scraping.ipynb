```python
# PUT THE DRIVER IN ANACONDA3 IN MONGED AND PUT IT IN SCRIPT FOLDER
```

```python
list = []
no = 0
for i in range(0,6):

    url = 'https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst&start='+str(i)
    browser = start_chrome(url)                                                     # Helium
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    
    for i in soup.find_all('div', class_ = 'css-1gatmva e1v1l3u10'):
        job_url = i.find_all('a',class_= 'css-o171kl')[0]['href']
        go_to('https://wuzzuf.net'+job_url)
        soup = BeautifulSoup(browser.page_source, 'html.parser')
        
        title = soup.find('h1', class_='css-f9uh36').text
        
         
        try :
            company_name = soup.find_all('a', class_='css-p7pghv')[0].text # there is company name
        except:
            company_name = soup.find_all('div', class_='css-9iujih')[0].text[:-2] #hidden company name
                
        try:
            company_headquarter = soup.find_all('span', class_='css-nhiaul')[0].text
        except:
            company_headquarter = 'unknown'
            
        try:
            company_url = soup.find_all('a', class_='css-p7pghv')[0]['href']
        except:
            company_url = 'unknown'
            
        location_of_job = soup.find_all('strong', class_='css-9geu3q')[0].text.split('-')[1][1:]
        
        try:
            company_type = soup.find_all('span', class_='css-xilyze')[0].text
        except:
            company_type = 'unknown'
        
        posted_time = soup.find_all('span', class_='css-182mrdn')[0].text
    
        
        collect_date =  time.strftime("%Y-%m-%d")
        
        section = soup.find_all('span', class_='css-wn0avc')
        
        for i in range(len(section)):
            if section[i].text[:-1] == 'Experience Needed':
                try:
                    Experience_Needed =  soup.find_all('span', class_='css-4xky9y')[i].text
                except:
                    Experience_Needed = 'not mentioned'
            elif section[i].text[:-1] == 'Career Level':
                try:
                    Career_Level = soup.find_all('span', class_='css-4xky9y')[i].text
                except:
                    Career_Level = 'not mentioned'
            elif section[i].text[:-1] == 'Education Level':
                try:
                    Education_Level = soup.find_all('span', class_='css-4xky9y')[i].text
                except:Education_Level = 'not mentioned'
            elif section[i].text[:-1] == 'Salary':
                try:
                    Salary = soup.find_all('span', class_='css-4xky9y')[i].text
                except:
                    Salary = 'not mentioned'
            else :
                pass
        
        
        job_categories = ''
        for i in soup.find_all('ul', class_='css-h5dsne')[0].find_all('span'):
            job_categories = job_categories + i.text + ' - '
        job_categories = job_categories[:-2]
        
        
        try:
        
            skills_and_tools = ''
            for i in soup.find_all('span', class_="css-6to1q"):
                skills_and_tools = skills_and_tools + i.text + ' - '
            skills_and_tools = skills_and_tools[:-2]
        except:
            skills_and_tools = 'not mentioned'
        
        try:
            Job_Description = ''
            for i in soup.find_all('div', class_='css-1uobp1k')[0].find_all('li'):
                Job_Description = Job_Description + i.text + ' - '
            Job_Description = Job_Description[:-2]
        except:
            Job_Description = 'not mentioned'
        
        try:
            Job_Requirements = ''
            for i in soup.find_all('div', class_='css-1t5f0fr')[0].find_all('li'):
                Job_Requirements = Job_Requirements + i.text + ' - '
            Job_Requirements[:-2]
        except:
            Job_Requirements = 'not mentioned'
        
        list.append({'role_title' :title, 
                     'company_name':company_name ,
                     'company_industry': company_type,
                     'company_headquarter': company_headquarter,
                     'company_url':company_url,
                     'location_of_job' : location_of_job,
                     'posted_time' : posted_time,
                     'collect_date' : collect_date,
                     'experience_needed': Experience_Needed,
                     'career_level': Career_Level,
                     'education_level': Education_Level,
                     'salary': Salary,
                     'job_categories': job_categories,
                     'skills_and_tools':skills_and_tools,
                     'job_description' :Job_Description,
                     'job_requirements':Job_Requirements
                     })
        time.sleep(3)
        no += 1
        print( str(no) + '-page compeleted')
    kill_browser()
```

```python
df = pd.DataFrame(list)
```
